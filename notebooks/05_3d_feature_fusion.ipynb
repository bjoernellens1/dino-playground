# 05 3D Feature Fusion

Fuse DINOv3 features into a 3D representation (TSDF/Voxel Grid).

```python
import torch
from dinov3_lab.tasks.fusion3d.tsdf import TSDFIntegrator

integrator = TSDFIntegrator(voxel_size=0.04)

# Dummy data
depth = torch.rand(480, 640)
pose = torch.eye(4)
intrinsics = torch.eye(3)

print("Integrating frame...")
integrator.integrate(depth, pose, intrinsics)
print("Done.")
```
